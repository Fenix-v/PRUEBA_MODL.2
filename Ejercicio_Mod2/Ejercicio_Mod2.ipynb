## Histograma y Curvas de Densidad

```python
def histograma(sub, var, la, mean, medi, maxi, mini, r, tit):
    plt.subplot(1, 2, sub)
    plt.hist(var, bins=50, color='blue', alpha=0.4)
    plt.xlabel("la")
    plt.ylabel("Cantidad viviendas")
    plt.axvline(mean, color='red', linestyle= 'dashed', linewidth=1, label=(f"Media: {mean:.2f} {r}"))
    plt.axvline(medi, color='green', linestyle= 'dashed', linewidth=1, label=(f"Mediana: {medi:.2f} {r}"))
    plt.axvline(maxi, color='black', linestyle= 'dashed', linewidth=1, label=(f"Maximo: {maxi:.2f} {r}"))
    plt.axvline(mini, color='orange', linestyle= 'dashed', linewidth=1, label=(f"Minimo: {mini:.2f} {r}"))
    plt.title(f"Histograma {tit}")
    plt.legend()
    plt.show
    
```

```python


plt.figure(figsize=(16, 4))
histograma(1, price, "Precio", mean_price, medi_price, max_price, min_price, "€", "Precios")
histograma(2, sqft, "Dimensiones", mean_sqft, medi_sqft, max_sqft, min_sqft, "m2", "m2")

plt.figure(figsize=(16, 4))
histograma(1, rooms, "Habitaciones", mean_rooms, medi_rooms, max_rooms, min_rooms, "", "Habitaciones")
histograma(2, baths, "Cuartos de Baño", mean_baths, medi_baths, max_baths, min_baths, "", "Cuartos de baño")


```

* CURVA DE DENSIDAD

```python
def curva_densidad(sub, var, xla, mean, medi, max, min, r, tit):
    plt.subplot(1, 2, sub)
    sns.kdeplot(var, label=xla)
    plt.axvline(mean, color='red', linestyle= 'dashed', linewidth=1, label=(f"Media: {mean:.2f} {r}"))
    plt.axvline(medi, color='green', linestyle= 'dashed', linewidth=1, label=(f"Mediana: {medi:.2f} {r}"))
    plt.axvline(max, color='black', linestyle= 'dashed', linewidth=1, label=(f"Maximo: {max:.2f} {r}"))
    plt.axvline(min, color='orange', linestyle= 'dashed', linewidth=1, label=(f"Minimo: {min:.2f} {r}"))
    plt.title(f"Curva densidad {tit}")
    plt.legend()
    
```

```python
plt.figure(figsize=(16, 4))
curva_densidad(1, price, "Precio", mean_price, medi_price, max_price, min_price, "€", "Precios")
curva_densidad(2, sqft, "Dimensiones", mean_sqft, medi_sqft, max_sqft, min_sqft, "m2", "m2")
plt.show()

plt.figure(figsize=(16, 4))
curva_densidad(1, rooms, "Habitaciones", mean_rooms, medi_rooms, max_rooms, min_rooms, "", "Habitaciones")
curva_densidad(2, baths, "Cuartos de Baño", mean_baths, medi_baths, max_baths, min_baths, "", "Cuartos de baño")
plt.show()
```


## Filtro de Outliers

```python

def filtrar_outliers(array, k, Q1, Q3, IQR):
    limite_inferior = Q1 - IQR * k
    limite_superior = Q3 + IQR * k
    array_sin_outliers = array[(array >= limite_inferior) & (array <= limite_superior)]
    return array_sin_outliers

```

```python
precio_sin_outliers = filtrar_outliers(price, 1.5, Q1_price, Q3_price, iqr_price)
hab_sin_outliers = filtrar_outliers(rooms, 1.5, Q1_room, Q3_room, iqr_room)
baños_sin_outliers = filtrar_outliers(baths, 1.5, Q1_bath, Q3_bath, iqr_bath)
m2_sin_outliers = filtrar_outliers(sqft, 1.5, Q1_m2, Q3_m2, iqr_m2)

```

## Histogramas Con y Sin Outliers

```python
def  histograma3 (sub, xla, xla_sin, tit):
    plt.subplot(1, 2, sub)
    outliers = list(set(xla) - set(xla_sin))
    plt.xlabel(tit)
    plt.ylabel(" Cantidad de viviendas ")
    plt.hist(xla_sin, bins= 50, color= "blue", alpha=0.5,  label= "Viviendas sin outliers")
    plt.hist(outliers, bins= 50, color= "red",  label= "Viviendas con outliers")
    plt.title(f"Histograma {tit}")
    plt.legend()
```

```python
plt.figure(figsize=(16, 4))
histograma3(1, price, precio_sin_outliers, "por Precio")
histograma3(2, sqft, m2_sin_outliers, "por m2")
plt.show()

plt.figure(figsize=(16, 4))
histograma3(1, rooms, hab_sin_outliers, " por Habitaciones")
histograma3(2, baths, baños_sin_outliers, " por Cuartos de baño")
plt.show()

```
